```python
## Task 1: Mark all launch sites on a map
# Extract unique launch site coordinates
launch_sites = spacex_df[['Launch Site', 'Lat', 'Long']].drop_duplicates()

# Center the map on the average coordinates
map_center = [launch_sites['Lat'].mean(), launch_sites['Long'].mean()]
site_map = folium.Map(location=map_center, zoom_start=4)

# Add markers for each launch site
for index, row in launch_sites.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=row['Launch Site'],
        icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
    ).add_to(site_map)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities

# Coordinates of launch site (e.g. CCAFS SLC-40)
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Coordinates of closest proximity point (e.g. coastline)
coastline_lat = 28.56367
coastline_lon = -80.57163

# Step 1: Calculate distance
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Step 2: Draw line between launch site and coastline
coordinates = [[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]]
folium.PolyLine(coordinates, color='blue').add_to(site_map)

# Step 3: Add marker at coastline with distance label
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_coastline),
    )
)
distance_marker.add_to(site_map)

# Show updated map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
